So what's happening here is range(5) creates a list 0,1,2,3,4 - of 5 items to iterate over. These are called i. For each of these, there is an empty list called n instantiated.
For each of these i, we create 'for  j in len(d)', which just consists of one object. So we will only have 5 objects n the first iteration, though we have 2 nested for-loops.
Within each of these 5 objects, we append 1 list type object consisting of 3 numerical values to the empty list n, which make use of the values of d, and our pre-set of s.
We do this with another 'append' step for another set of numerical values. So each n is a set of 2 lists where *the middle element of the lists is the same, and the other 2 elements are just the negative of each other*.

```python
from math import sin, cos
from matplotlib import pyplot as plt 

initial_y = 0
initial_x = 0
scale = 1
spread = 0.6
branch_levels = 5
angle_of_rotation = 0
angle_of_increment = 0.1

d = [[initial_x, initial_y+scale, angle_of_rotation]]

plt.plot([initial_x,initial_x],[initial_y,initial_y+scale]) # plots trunk

s = scale
inc = angle_of_increment

for branch_level in range(branch_levels):
    n = []
    for fork in range(len(d)):
        x = d[fork][0]
        y = d[fork][1]
        z = d[fork][2]
        n.append([x+s*sin(z-inc), y+s*cos(z-inc), z-inc])
        n.append([x+s*sin(z+inc), y+s*cos(z+inc), z+inc])
        plt.plot([x, n[-2][0]], [y, n[-2][1]])
        plt.plot([x, n[-1][0]], [y, n[-1][1]])
    d = n
    s *= spread
    
plt.savefig('tree.png')
```

```python
from math import sin, cos
from matplotlib import pyplot as plt


def plotTree(position, scale, spread, branch_levels, rotation, increment):

    initial_x, initial_y = position
    s = scale
    inc = increment
    d = [[initial_x, initial_y+s, rotation]]

    plt.plot([initial_x, initial_x], [initial_y, initial_y+s])  # plot trunk

    for branch_level in range(branch_levels):
        n = []
        for i, fork in enumerate(d):
            a, b, c = d[i][0], d[i][1], d[i][2]
            n.append([a+s*sin(c-inc), b+s*cos(c-inc), c-inc])
            n.append([a+s*sin(c+inc), b+s*cos(c+inc), c+inc])
            plt.plot([a, n[-2][0]], [b, n[-2][1]])
            plt.plot([a, n[-1][0]], [b, n[-1][1]])
        d = n
        s *= spread

    plt.savefig('tree.png')


plotTree((0, 0), 1, 0.6, 5, 0, 0.1)

```
