```python
from math import sin, cos
from matplotlib import pyplot as plt


def plotTree(position, scale, branch_lengths, branch_levels, angle):

    x1, y1 = position
    rotation = 0
    d = [[x1, y1+scale, rotation]]

    plt.plot([x1, x1], [y1, y1+scale])  # plot trunk

    for branch_level in range(branch_levels):
        n = []
        for i, fork in enumerate(d):
            a, b, c = d[i][0], d[i][1], d[i][2]
            n.append([a+scale*sin(c-angle), b+scale*cos(c-angle), c-angle])
            n.append([a+scale*sin(c+angle), b+scale*cos(c+angle), c+angle])
            plt.plot([a, n[-2][0]], [b, n[-2][1]])
            plt.plot([a, n[-1][0]], [b, n[-1][1]])
        d = n
        scale *= branch_lengths

    plt.savefig('tree.png')


plotTree((0, 0), 1, 0.6, 5, 0.1)
```

```python
def plotTree(x1, y1, lengths, levels, angle):

    scale = 1
    rotation = 0
    d = [[x1, y1+scale, rotation]]

    plt.plot([x1, x1], [y1, y1+scale])  # plot trunk

    for level in range(levels):
        n = []
        for i, fork in enumerate(d):
            a, b, c = d[i][0], d[i][1], d[i][2]
            n.append([a+scale*sin(c-angle), b+scale*cos(c-angle), c-angle])
            n.append([a+scale*sin(c+angle), b+scale*cos(c+angle), c+angle])
            print("level:", level, "fork:", i, n)
            plt.plot([a, n[-2][0]], [b, n[-2][1]])
            plt.plot([a, n[-1][0]], [b, n[-1][1]])
        d = n
        scale *= lengths

    plt.savefig('tree.png')
    
plotTree(0, 0, 0.6, 5, 0.1)
```
